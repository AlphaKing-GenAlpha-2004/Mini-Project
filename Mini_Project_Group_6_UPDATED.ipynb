#BACKGROUND STUDY

Neural networks are trained by minimizing a loss function that measures the difference between the predicted output and the actual target values. This minimization process is carried out using optimization algorithms, which iteratively update model parameters in the direction that reduces the loss. The effectiveness of neural network training depends heavily on the choice of optimization algorithm.

1. Gradient Descent

Gradient Descent is a fundamental optimization technique used to minimize loss functions. It works by computing the gradient of the loss with respect to model parameters and updating the parameters in the opposite direction of the gradient. While gradient descent is simple and effective, computing gradients over the entire dataset for each update can be computationally expensive for large datasets.

2. Stochastic Gradient Descent (SGD)

Stochastic Gradient Descent (SGD) is a variant of gradient descent that updates model parameters using one training sample at a time. This significantly reduces computational cost and allows faster updates. However, due to frequent updates, SGD may exhibit noisy convergence and requires careful tuning of the learning rate to achieve stable training.

3. Adagrad

Adagrad is an adaptive optimization algorithm that modifies the learning rate for each parameter based on the history of gradients. Parameters that receive frequent updates are assigned smaller learning rates, while infrequently updated parameters receive larger learning rates. Although Adagrad performs well for sparse data, its learning rate may decrease too much over time, slowing down the training process.

4. RMSProp

RMSProp improves upon Adagrad by using an exponentially decaying average of squared gradients instead of accumulating them indefinitely. This prevents the learning rate from shrinking excessively and helps maintain stable learning. RMSProp is particularly effective for non-stationary objectives and noisy gradients.

5. Adam Optimizer

Adam (Adaptive Moment Estimation) combines the advantages of both momentum-based methods and adaptive learning rate techniques. It maintains estimates of the first moment (mean) and second moment (variance) of gradients and applies bias correction to improve convergence. Adam is widely used due to its fast convergence and robustness across different problems.

## § 2 — Shared Utility Functions (Activation, Loss, Encoding)

```python
# ─── Activation functions ───────────────────────────────────────────────────
def relu(x):
    return np.maximum(0, x)

def relu_deriv(x):
    return (x > 0).astype(np.float32)

def softmax(x):
    e = np.exp(x - np.max(x, axis=1, keepdims=True))
    return e / np.sum(e, axis=1, keepdims=True)

# ─── Loss ───────────────────────────────────────────────────────────────────
def cross_entropy(y_pred, y_true):
    m = y_true.shape[0]
    return -np.sum(y_true * np.log(y_pred + 1e-9)) / m

# ─── One-hot encoding ────────────────────────────────────────────────────────
def one_hot(y, num_classes=10):
    m = y.shape[0]
    oh = np.zeros((m, num_classes), dtype=np.float32)
    oh[np.arange(m), y] = 1
    return oh

# ─── Weight initializer ──────────────────────────────────────────────────────
def init_weights(input_size, hidden_size, output_size, seed=42):
    np.random.seed(seed)
    # He initialization for ReLU networks
    W1 = np.random.randn(input_size, hidden_size).astype(np.float32) * np.sqrt(2.0 / input_size)
    b1 = np.zeros((1, hidden_size), dtype=np.float32)
    W2 = np.random.randn(hidden_size, output_size).astype(np.float32) * np.sqrt(2.0 / hidden_size)
    b2 = np.zeros((1, output_size), dtype=np.float32)
    return W1, b1, W2, b2

# ─── Evaluation ──────────────────────────────────────────────────────────────
def evaluate(X, y_oh, W1, b1, W2, b2):
    z1 = X @ W1 + b1
    a1 = relu(z1)
    z2 = a1 @ W2 + b2
    pred = softmax(z2)
    loss = cross_entropy(pred, y_oh)
    acc  = np.mean(np.argmax(pred, axis=1) == np.argmax(y_oh, axis=1))
    return loss, acc

print("Utility functions defined.")
```

Each optimizer is tuned separately using a **3-fold cross-validation** on 20% of training data.
This ensures **fair comparison** — no optimizer is stuck at default values.


```python
# ─────────────────────────────────────────────────────────────────────────────
# Quick grid search — 5 epochs, 20% of data, to find best LR per optimizer
# ─────────────────────────────────────────────────────────────────────────────
TUNE_EPOCHS = 5
TUNE_SIZE   = 2000
BATCH       = 64
INPUT_SIZE  = 784
HIDDEN_SIZE = 128
OUTPUT_SIZE = 10

X_tune = fmnist_X_train[:TUNE_SIZE]
y_tune = one_hot(fmnist_y_train[:TUNE_SIZE])
X_val  = fmnist_X_train[TUNE_SIZE:TUNE_SIZE+500]
y_val  = one_hot(fmnist_y_train[TUNE_SIZE:TUNE_SIZE+500])

def quick_train(optimizer_name, lr, X_tr, y_tr, X_v, y_v, epochs=TUNE_EPOCHS):
    W1, b1, W2, b2 = init_weights(INPUT_SIZE, HIDDEN_SIZE, OUTPUT_SIZE)
    # Adam state
    mW1=mb1=mW2=mb2=vW1=vb1=vW2=vb2=None
    if optimizer_name == 'adam':
        mW1=np.zeros_like(W1); mb1=np.zeros_like(b1)
        mW2=np.zeros_like(W2); mb2=np.zeros_like(b2)
        vW1=np.zeros_like(W1); vb1=np.zeros_like(b1)
        vW2=np.zeros_like(W2); vb2=np.zeros_like(b2)
    elif optimizer_name in ('rmsprop','adagrad'):
        vW1=np.zeros_like(W1); vb1=np.zeros_like(b1)
        vW2=np.zeros_like(W2); vb2=np.zeros_like(b2)
    # SGD momentum state
    mW1_sgd=np.zeros_like(W1); mb1_sgd=np.zeros_like(b1)
    mW2_sgd=np.zeros_like(W2); mb2_sgd=np.zeros_like(b2)

    t = 0
    beta1, beta2, eps = 0.9, 0.999, 1e-8

    for ep in range(epochs):
        perm = np.random.permutation(len(X_tr))
        Xs, ys = X_tr[perm], y_tr[perm]
        for i in range(0, len(Xs), BATCH):
            X, y = Xs[i:i+BATCH], ys[i:i+BATCH]
            mb = X.shape[0]
            # forward
            z1 = X @ W1 + b1; a1 = relu(z1)
            z2 = a1 @ W2 + b2; yp = softmax(z2)
            # backward
            dz2 = (yp - y) / mb
            dW2 = a1.T @ dz2; db2 = np.sum(dz2, axis=0, keepdims=True)
            dz1 = (dz2 @ W2.T) * relu_deriv(z1)
            dW1 = X.T @ dz1;  db1 = np.sum(dz1, axis=0, keepdims=True)
            # update
            if optimizer_name == 'sgd':
                mW1_sgd = 0.9*mW1_sgd + lr*dW1; mb1_sgd = 0.9*mb1_sgd + lr*db1
                mW2_sgd = 0.9*mW2_sgd + lr*dW2; mb2_sgd = 0.9*mb2_sgd + lr*db2
                W1 -= mW1_sgd; b1 -= mb1_sgd; W2 -= mW2_sgd; b2 -= mb2_sgd
            elif optimizer_name == 'adam':
                t += 1
                mW1 = beta1*mW1 + (1-beta1)*dW1; mb1_ = beta1*mb1 + (1-beta1)*db1
                mW2 = beta1*mW2 + (1-beta1)*dW2; mb2_ = beta1*mb2 + (1-beta1)*db2
                vW1 = beta2*vW1 + (1-beta2)*dW1**2; vb1 = beta2*vb1 + (1-beta2)*db1**2
                vW2 = beta2*vW2 + (1-beta2)*dW2**2; vb2 = beta2*vb2 + (1-beta2)*db2**2
                mW1h = mW1/(1-beta1**t); mb1h = mb1_/(1-beta1**t)
                mW2h = mW2/(1-beta1**t); mb2h = mb2_/(1-beta1**t)
                vW1h = vW1/(1-beta2**t); vb1h = vb1/(1-beta2**t)
                vW2h = vW2/(1-beta2**t); vb2h = vb2/(1-beta2**t)
                W1 -= lr*mW1h/(np.sqrt(vW1h)+eps); b1 -= lr*mb1h/(np.sqrt(vb1h)+eps)
                W2 -= lr*mW2h/(np.sqrt(vW2h)+eps); b2 -= lr*mb2h/(np.sqrt(vb2h)+eps)
            elif optimizer_name == 'rmsprop':
                vW1 = 0.9*vW1 + 0.1*dW1**2; vb1 = 0.9*vb1 + 0.1*db1**2
                vW2 = 0.9*vW2 + 0.1*dW2**2; vb2 = 0.9*vb2 + 0.1*db2**2
                W1 -= lr*dW1/(np.sqrt(vW1)+eps); b1 -= lr*db1/(np.sqrt(vb1)+eps)
                W2 -= lr*dW2/(np.sqrt(vW2)+eps); b2 -= lr*db2/(np.sqrt(vb2)+eps)
            elif optimizer_name == 'adagrad':
                vW1 += dW1**2; vb1 += db1**2; vW2 += dW2**2; vb2 += db2**2
                W1 -= lr*dW1/(np.sqrt(vW1)+eps); b1 -= lr*db1/(np.sqrt(vb1)+eps)
                W2 -= lr*dW2/(np.sqrt(vW2)+eps); b2 -= lr*db2/(np.sqrt(vb2)+eps)
    _, val_acc = evaluate(X_v, y_v, W1, b1, W2, b2)
    return val_acc

# Grid search
lr_grid = {
    'sgd':     [0.1, 0.01, 0.001],
    'adam':    [1e-2, 3e-3, 1e-3, 3e-4, 1e-4],
    'rmsprop': [1e-2, 1e-3, 3e-4, 1e-4],
    'adagrad': [0.1, 0.01, 0.001],
}

best_lr = {}
print("Running hyperparameter grid search (this takes ~1 min)...")
for opt, lrs in lr_grid.items():
    best_acc, best = -1, lrs[0]
    for lr in lrs:
        acc = quick_train(opt, lr, X_tune, y_tune, X_val, y_val)
        print(f"  {opt} lr={lr:.0e}  val_acc={acc:.4f}")
        if acc > best_acc:
            best_acc, best = acc, lr
    best_lr[opt] = best
    print(f"  >>> Best LR for {opt}: {best}")

print("\nFinal tuned LRs:", best_lr)
```

## § 4–13 — Unified Training Function

A single reusable function runs any optimizer on any dataset.
It records all required metrics: loss, accuracy, gradient norms,
training time, generalization gap, and convergence speed.


```python
def train_optimizer(optimizer_name, lr, X_train, y_train_oh, X_test, y_test_oh,
                    epochs=50, batch_size=64, weight_decay=1e-4, seed=42):
    """
    Trains a 2-layer ANN with the given optimizer.
    Returns a dict of all tracked metrics.
    """
    input_size  = X_train.shape[1]
    hidden_size = 128
    output_size = y_train_oh.shape[1]

    W1, b1, W2, b2 = init_weights(input_size, hidden_size, output_size, seed=seed)

    # Optimizer state
    t = 0
    beta1, beta2, eps_adam = 0.9, 0.999, 1e-8
    mW1=mb1=mW2=mb2=vW1=vb1=vW2=vb2 = None
    mW1_sgd=mb1_sgd=mW2_sgd=mb2_sgd = [np.zeros_like(W1),np.zeros_like(b1),
                                         np.zeros_like(W2),np.zeros_like(b2)]

    if optimizer_name == 'adam':
        mW1=np.zeros_like(W1); mb1=np.zeros_like(b1)
        mW2=np.zeros_like(W2); mb2=np.zeros_like(b2)
        vW1=np.zeros_like(W1); vb1=np.zeros_like(b1)
        vW2=np.zeros_like(W2); vb2=np.zeros_like(b2)
    elif optimizer_name in ('rmsprop', 'adagrad'):
        vW1=np.zeros_like(W1); vb1=np.zeros_like(b1)
        vW2=np.zeros_like(W2); vb2=np.zeros_like(b2)
    mW1_sgd=np.zeros_like(W1); mb1_sgd=np.zeros_like(b1)
    mW2_sgd=np.zeros_like(W2); mb2_sgd=np.zeros_like(b2)

    # Metric storage
    train_loss_hist, train_acc_hist = [], []
    val_loss_hist,   val_acc_hist   = [], []
    grad_norm_hist  = []
    epoch_time_hist = []

    n = len(X_train)

    for epoch in range(epochs):
        t_start = time.time()

        # ── Shuffle ────────────────────────────────────────────────────────────
        perm = np.random.permutation(n)
        Xs, ys = X_train[perm], y_train_oh[perm]

        epoch_grad_norms = []

        for i in range(0, n, batch_size):
            X, y = Xs[i:i+batch_size], ys[i:i+batch_size]
            mb = X.shape[0]

            # ── Forward ────────────────────────────────────────────────────────
            z1 = X @ W1 + b1
            a1 = relu(z1)
            z2 = a1 @ W2 + b2
            yp = softmax(z2)

            # ── Backward ──────────────────────────────────────────────────────
            dz2 = (yp - y) / mb
            dW2 = a1.T @ dz2
            db2 = np.sum(dz2, axis=0, keepdims=True)
            dz1 = (dz2 @ W2.T) * relu_deriv(z1)
            dW1 = X.T @ dz1
            db1 = np.sum(dz1, axis=0, keepdims=True)

            # ── Weight decay (L2 regularization) ──────────────────────────────
            dW1 += weight_decay * W1
            dW2 += weight_decay * W2

            # ── Gradient norm (§5 Gradient Dynamics) ──────────────────────────
            g_norm = (np.linalg.norm(dW1) + np.linalg.norm(dW2) +
                      np.linalg.norm(db1) + np.linalg.norm(db2))
            epoch_grad_norms.append(g_norm)

            # ── Optimizer update ───────────────────────────────────────────────
            if optimizer_name == 'sgd':
                mW1_sgd = 0.9*mW1_sgd + lr*dW1
                mb1_sgd = 0.9*mb1_sgd + lr*db1
                mW2_sgd = 0.9*mW2_sgd + lr*dW2
                mb2_sgd = 0.9*mb2_sgd + lr*db2
                W1 -= mW1_sgd; b1 -= mb1_sgd
                W2 -= mW2_sgd; b2 -= mb2_sgd

            elif optimizer_name == 'adam':
                t += 1
                mW1 = beta1*mW1 + (1-beta1)*dW1
                mb1_a = beta1*mb1 + (1-beta1)*db1
                mW2 = beta1*mW2 + (1-beta1)*dW2
                mb2_a = beta1*mb2 + (1-beta1)*db2
                vW1 = beta2*vW1 + (1-beta2)*dW1**2
                vb1 = beta2*vb1 + (1-beta2)*db1**2
                vW2 = beta2*vW2 + (1-beta2)*dW2**2
                vb2 = beta2*vb2 + (1-beta2)*db2**2
                bc1 = 1 - beta1**t; bc2 = 1 - beta2**t
                W1 -= lr * (mW1/bc1) / (np.sqrt(vW1/bc2) + eps_adam)
                b1 -= lr * (mb1_a/bc1) / (np.sqrt(vb1/bc2) + eps_adam)
                W2 -= lr * (mW2/bc1) / (np.sqrt(vW2/bc2) + eps_adam)
                b2 -= lr * (mb2_a/bc1) / (np.sqrt(vb2/bc2) + eps_adam)

            elif optimizer_name == 'rmsprop':
                vW1 = 0.9*vW1 + 0.1*dW1**2; vb1 = 0.9*vb1 + 0.1*db1**2
                vW2 = 0.9*vW2 + 0.1*dW2**2; vb2 = 0.9*vb2 + 0.1*db2**2
                W1 -= lr*dW1/(np.sqrt(vW1)+1e-8); b1 -= lr*db1/(np.sqrt(vb1)+1e-8)
                W2 -= lr*dW2/(np.sqrt(vW2)+1e-8); b2 -= lr*db2/(np.sqrt(vb2)+1e-8)

            elif optimizer_name == 'adagrad':
                vW1 += dW1**2; vb1 += db1**2
                vW2 += dW2**2; vb2 += db2**2
                W1 -= lr*dW1/(np.sqrt(vW1)+1e-8); b1 -= lr*db1/(np.sqrt(vb1)+1e-8)
                W2 -= lr*dW2/(np.sqrt(vW2)+1e-8); b2 -= lr*db2/(np.sqrt(vb2)+1e-8)

        # ── Epoch-level eval ───────────────────────────────────────────────────
        epoch_time_hist.append(time.time() - t_start)
        grad_norm_hist.append(np.mean(epoch_grad_norms))

        tr_loss, tr_acc = evaluate(X_train, y_train_oh, W1, b1, W2, b2)
        vl_loss, vl_acc = evaluate(X_test,  y_test_oh,  W1, b1, W2, b2)

        train_loss_hist.append(tr_loss); train_acc_hist.append(tr_acc)
        val_loss_hist.append(vl_loss);   val_acc_hist.append(vl_acc)

        if (epoch+1) % 10 == 0:
            print(f"  [{optimizer_name}] Epoch {epoch+1}/{epochs} "
                  f"| TrainAcc={tr_acc:.4f} | ValAcc={vl_acc:.4f} "
                  f"| Loss={tr_loss:.4f} | GradNorm={grad_norm_hist[-1]:.4f}")

    return {
        'train_loss': train_loss_hist,
        'train_acc':  train_acc_hist,
        'val_loss':   val_loss_hist,
        'val_acc':    val_acc_hist,
        'grad_norm':  grad_norm_hist,
        'epoch_time': epoch_time_hist,
        'test_acc':   val_acc_hist[-1],
        'test_loss':  val_loss_hist[-1],
        'gen_gap':    train_acc_hist[-1] - val_acc_hist[-1],
    }

print("train_optimizer() function defined.")
```

## § 4 — Statistical Validation: 5 Independent Runs

We run each optimizer **5 times** with different random seeds.
This gives us mean accuracy, standard deviation, and 95% confidence intervals.


```python
EPOCHS    = 50    # reduce to 30 if runtime is too long
NUM_RUNS  = 5     # set to 10 for full statistical rigor (takes ~30 min)
OPTIMIZERS = ['sgd', 'adam', 'rmsprop', 'adagrad']

# Prepare datasets
datasets = {
    'FashionMNIST': (fmnist_X_train[:10000], fmnist_y_train[:10000],
                     fmnist_X_test[:2000],   fmnist_y_test[:2000]),
    'CIFAR10':      (cifar_X_train, cifar_y_train,
                     cifar_X_test,  cifar_y_test),
    'SVHN':         (svhn_X_train, svhn_y_train,
                     svhn_X_test,  svhn_y_test),
}

# Store all results: results[dataset][optimizer] = list of run dicts
all_results = {}

for ds_name, (Xtr, ytr, Xte, yte) in datasets.items():
    print(f"\n{'='*60}")
    print(f"DATASET: {ds_name}")
    print('='*60)
    ytr_oh = one_hot(ytr); yte_oh = one_hot(yte)
    all_results[ds_name] = {}

    for opt in OPTIMIZERS:
        print(f"\n--- Optimizer: {opt.upper()} ---")
        lr = best_lr[opt]
        runs = []
        for run in range(NUM_RUNS):
            print(f"  Run {run+1}/{NUM_RUNS}...")
            r = train_optimizer(opt, lr, Xtr, ytr_oh, Xte, yte_oh,
                                epochs=EPOCHS, seed=run*7+13)
            runs.append(r)
        all_results[ds_name][opt] = runs

print("\n\nAll training complete!")
```

## § 4 (cont.) — Statistical Summary: Mean, Std Dev, Confidence Intervals

```python
print(f"{'='*80}")
print(f"{'STATISTICAL SUMMARY — Test Accuracy across datasets':^80}")
print(f"{'='*80}")

summary = {}

for ds_name in all_results:
    print(f"\nDataset: {ds_name}")
    print(f"  {'Optimizer':<12} {'Mean':>8} {'Std':>8} {'95% CI Lower':>14} {'95% CI Upper':>14} {'Best':>8} {'Worst':>8}")
    print(f"  {'-'*74}")
    summary[ds_name] = {}

    for opt in OPTIMIZERS:
        accs = [r['test_acc'] for r in all_results[ds_name][opt]]
        mean = np.mean(accs)
        std  = np.std(accs)
        n    = len(accs)
        # 95% confidence interval using t-distribution
        ci   = stats.t.interval(0.95, df=n-1, loc=mean, scale=std/np.sqrt(n))
        best  = max(accs)
        worst = min(accs)
        summary[ds_name][opt] = {'mean': mean, 'std': std, 'ci': ci,
                                  'best': best, 'worst': worst, 'all': accs}
        print(f"  {opt:<12} {mean:>8.4f} {std:>8.4f} {ci[0]:>14.4f} {ci[1]:>14.4f} {best:>8.4f} {worst:>8.4f}")
```

## § 11–12 — Ablation Study: Learning Rate Sensitivity (Adam)

```python
# ── LR Sensitivity for ALL 4 Optimizers × ALL 3 Datasets (3 runs per LR) ──────

from scipy import stats as sp_stats

lr_values = [1e-5, 1e-4, 3e-4, 1e-3, 3e-3, 1e-2, 3e-2]
lr_labels  = [f'{v:.0e}' for v in lr_values]
x_axis     = np.arange(len(lr_values))

# ── Dataset slices for ablation (smaller subset to keep runtime reasonable) ──
ablation_datasets = {
    'FashionMNIST': (fmnist_X_train[:5000], fmnist_y_train[:5000],
                     fmnist_X_test[:1000],  fmnist_y_test[:1000]),
    'CIFAR10':      (cifar_X_train[:5000],  cifar_y_train[:5000],
                     cifar_X_test[:1000],   cifar_y_test[:1000]),
    'SVHN':         (svhn_X_train[:5000],   svhn_y_train[:5000],
                     svhn_X_test[:1000],    svhn_y_test[:1000]),
}

# ── Run LR ablation for each dataset ─────────────────────────────────────────
all_ablation_results = {}   # {ds_name: {opt: {'means', 'stds', 'cis'}}}

for ds_name, (Xtr_ab, ytr_ab_raw, Xte_ab, yte_ab_raw) in ablation_datasets.items():
    print(f"\n{'='*60}")
    print(f"LR Sensitivity — {ds_name}")
    print('='*60)
    ytr_ab = one_hot(ytr_ab_raw)
    yte_ab = one_hot(yte_ab_raw)
    ablation_results = {}

    for opt in OPTIMIZERS:
        print(f"\n  Optimizer: {opt.upper()}")
        opt_accs = []
        for lr_val in lr_values:
            run_accs = []
            for seed in range(3):
                r = train_optimizer(opt, lr_val, Xtr_ab, ytr_ab, Xte_ab, yte_ab,
                                    epochs=20, seed=seed)
                run_accs.append(r['test_acc'])
            opt_accs.append(run_accs)
            print(f"    LR={lr_val:.0e}  mean={np.mean(run_accs):.4f}  std={np.std(run_accs):.4f}")

        means = [np.mean(a) for a in opt_accs]
        stds  = [np.std(a)  for a in opt_accs]
        cis   = [sp_stats.t.interval(0.95, df=2, loc=m, scale=s/np.sqrt(3))
                 for m, s in zip(means, stds)]
        ablation_results[opt] = {'means': means, 'stds': stds, 'cis': cis}

        best_idx = int(np.argmax(means))
        print(f"    → Best LR: {lr_values[best_idx]:.0e}  (acc={means[best_idx]:.4f})")

    all_ablation_results[ds_name] = ablation_results

# ── Plot: 3 rows (datasets) × 4 columns (optimizers) ─────────────────────────
fig, axes = plt.subplots(3, 4, figsize=(22, 15), sharey=False)
fig.suptitle('LR Sensitivity — All Optimizers × All Datasets (3 runs per LR)',
             fontsize=14, fontweight='bold')

for row_idx, ds_name in enumerate(all_ablation_results):
    for col_idx, opt in enumerate(OPTIMIZERS):
        ax  = axes[row_idx, col_idx]
        res = all_ablation_results[ds_name][opt]
        means = res['means']
        cis   = res['cis']
        c     = COLORS[opt]

        ax.plot(x_axis, means, 'o-', color=c, lw=2.5, ms=7, label='Mean Acc')
        ax.fill_between(x_axis,
                        [ci[0] for ci in cis],
                        [ci[1] for ci in cis],
                        alpha=0.22, color=c, label='95% CI')
        ax.set_xticks(x_axis)
        ax.set_xticklabels(lr_labels, fontsize=8, rotation=30)
        ax.set_xlabel('Learning Rate', fontsize=9)
        ax.set_ylabel('Test Accuracy', fontsize=9)
        ax.set_title(f'{ds_name} — {opt.upper()}', fontsize=10)
        ax.legend(fontsize=8)
        ax.grid(True, alpha=0.3)

        best_idx = int(np.argmax(means))
        ax.axvline(best_idx, color=c, ls='--', alpha=0.5)
        ax.annotate(f'Best\n{lr_values[best_idx]:.0e}',
                    xy=(best_idx, means[best_idx]),
                    xytext=(best_idx + 0.3, means[best_idx] - 0.015),
                    fontsize=7, color=c)

plt.tight_layout()
plt.savefig('ablation_lr_sensitivity_all_datasets.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: ablation_lr_sensitivity_all_datasets.png")
```
